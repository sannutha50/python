**What is the total number of business trips versus personal trips?**

```python
total=uber_data['CATEGORY*'].value_counts()
print(total)
```

**What percentage of trips are business versus personal?**

```python
trip_count=uber_data['CATEGORY*'].count()
total_trips=uber_data['CATEGORY*'].value_counts()
total_percent=(total_trips/trip_count)*100
print(total_percent)
```
